```python
def missing_values_table(dataframe, plot=False):
    nan_columns = [col for col in dataframe.columns if dataframe[col].isnull().sum() > 0]
    n_miss = dataframe[nan_columns].isnull().sum().sort_values(ascending=False)
    ratio = (dataframe[nan_columns].isnull().sum() / dataframe.shape[0] * 100).sort_values(ascending=False)
    missing_df = pd.concat([n_miss, np.round(ratio, 2)], axis=1, keys=['n_miss', 'ratio'])
    
    if plot:
        plt.figure(figsize=(10, 8))
        bars = plt.barh(missing_df.index, missing_df['ratio'], color=sns.color_palette('pastel'))  
        plt.ylabel('Features')  
        plt.xlabel('Percentage of Missing Values')  
        plt.title('Missing Values by Feature')  
        
        for bar in bars:
            xval = bar.get_width()  
            plt.text(xval, bar.get_y() + bar.get_height()/2, f'{xval:.2f}%', va='center', ha='left', fontsize=10) 
        
        plt.xticks(fontsize=8) 
        plt.yticks(fontsize=8) 
        plt.grid(True)  
        plt.tight_layout()
        plt.show()
    else:
        return missing_df
```

```python
nan_columns = missing_values_table(df, plot=True)
```

```python
folder_path = '/kaggle/input/rsna-2024-lumbar-spine-degenerative-classification/train_images/100206310/1012284084'
dicom_files = [f for f in os.listdir(folder_path) if f.endswith('.dcm')]
```

```python
study_id = int(folder_path.split('/')[-2])

study_label_coordinates = label_coordinates_df[label_coordinates_df['study_id'] == study_id]

# Initialize lists to store filtered DICOM files and corresponding coordinates
filtered_dicom_files = []
filtered_label_coordinates = []

# Iterate over DICOM files to filter those with corresponding label coordinates
for dicom_file in dicom_files:
    instance_number = int(dicom_file.split('.')[0])
    corresponding_coordinates = study_label_coordinates[study_label_coordinates['instance_number'] == instance_number]
    if not corresponding_coordinates.empty:
        filtered_dicom_files.append(dicom_file)
        filtered_label_coordinates.append(corresponding_coordinates)


num_images = len(filtered_dicom_files)
num_rows = (num_images // 3) + (1 if num_images % 3 != 0 else 0)
num_cols = min(num_images, 3)


fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))

for i, (dicom_file, label_coordinates) in enumerate(zip(filtered_dicom_files, filtered_label_coordinates)):
    row_index = i // 3
    col_index = i % 3
    
    dicom_file_path = os.path.join(folder_path, dicom_file)
    dicom_data = pydicom.dcmread(dicom_file_path)
    image = dicom_data.pixel_array   
    axs[row_index, col_index].imshow(image, cmap='gray')
    axs[row_index, col_index].set_title(f'DICOM Image - {dicom_file}')
    axs[row_index, col_index].axis('off')   
    for _, row in label_coordinates.iterrows():
        axs[row_index, col_index].plot(row['x'], row['y'], 'ro', markersize=5) 

# Hide empty subplots if num_images is not a multiple of 3
for j in range(num_images, num_rows * 3):
    row_index = j // 3
    col_index = j % 3
    axs[row_index, col_index].axis('off')

plt.tight_layout()
plt.show();
```

<div style="text-align:center; font-size:20px; border: 2px solid #008080; border-radius: 5px; padding: 8px; margin-bottom: 10px; background-color:white ; font-family: Montserrat;font-size: 200%;font-weight: bold; color: #008080;">Data Preparation and Preprocessing
</div>

```python
df_melted = df.melt(id_vars=['study_id'], var_name='condition_level', value_name='severity')
df_melted[['condition', 'level']] = df_melted['condition_level'].str.rsplit('_', n=1, expand=True)

# Fill missing values with mode
fill = df_melted['severity'].mode()[0]
df_melted['severity'].fillna(fill, inplace=True)

# Convert Severity variable to numeric values using LabelEncoder
le = LabelEncoder()
df_melted['severity_encoded'] = le.fit_transform(df_melted['severity'])

# Separate dependent and independent variables
x_train = df_melted[['study_id', 'condition', 'level']]
y_train = df_melted['severity_encoded']

# Apply one-hot encoding
x_train = pd.get_dummies(x_train, columns=['condition', 'level']).astype(int)
```

<div style="text-align:center; font-size:20px; border: 2px solid #008080; border-radius: 5px; padding: 8px; margin-bottom: 10px; background-color:white ; font-family: Montserrat; font-size: 200%;font-weight: bold; color: #008080;">Test Data Set Preparation and Preprocessing
</div>

```python
## Predictions on test data
test_rows = []
for _, row in test_series.iterrows():
    for condition in ['left_neural_foraminal_narrowing', 'right_neural_foraminal_narrowing', 'left_subarticular_stenosis', 'right_subarticular_stenosis', 'spinal_canal_stenosis']:
        for level in ['l1_l2', 'l2_l3', 'l3_l4', 'l4_l5', 'l5_s1']:
            test_rows.append({
                'study_id': row['study_id'],
                'condition': condition,
                'level': level
            })

x_test = pd.DataFrame(test_rows)

# One-hot encoding for test data
x_test= pd.get_dummies(x_test, columns=['condition', 'level']).astype(int)

# Align columns with training data and fill missing columns with 0
x_test= x_test.reindex(columns=x_train.columns, fill_value=0)
```

<div style="text-align:center; font-size:20px; border: 2px solid #008080; border-radius: 5px; padding: 8px; margin-bottom: 10px; background-color:white ; font-family: Montserrat; font-size: 200%; font-weight: bold; color: #008080;">Model Training and Prediction
</div>

```python
# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, test_size=0.3, random_state=17)
```

```python
# Create and train the RandomForestClassifier model
rf_model = RandomForestClassifier(random_state=46)

rf_model.fit(x_train, y_train)
```

```python
# Now predict on x_test and evaluate metrics
y_pred = rf_model.predict(x_test)

# Calculate performance metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print("RandomForestClassifier Metrics:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")
```

```python
# Predictions
predictions_proba = rf_model.predict_proba(x_test)

# Organize results in a data frame
predictions_df = pd.DataFrame(predictions_proba, columns=le.classes_)
predictions_df['study_id'] = x_test['study_id'].values

# Create condition_level using zip function
predictions_df['condition_level'] = df_melted.apply(lambda row: f"{row['condition']}_{row['level']}", axis=1)
predictions_df['row_id'] = predictions_df['study_id'].astype(str) + '_' + predictions_df['condition_level']

# Calculating Normal/Mild, Moderate, Severe ratios
predictions_df['normal_mild'] = predictions_df['Normal/Mild'] / 2.3
predictions_df['moderate'] = predictions_df['Moderate'] * 1.47
predictions_df['severe'] = 1 - (predictions_df['normal_mild'] + predictions_df['moderate'])

# Drop unnecessary columns
drop_list = ["Normal/Mild", "Moderate", "Severe", "study_id", "condition_level"]
predictions_df.drop(drop_list, axis=1, inplace=True)

# Show results
predictions_df.head(15)
```

```python
# Set the style and color palette
plt.figure(figsize=(8, 6))
sns.set(style='whitegrid', palette='Set2')

# Density plot for Normal/Mild, Moderate, and Severe
sns.kdeplot(predictions_df['normal_mild'], shade=True, label='Normal/Mild', alpha=0.8, linewidth=2)
sns.kdeplot(predictions_df['moderate'], shade=True, label='Moderate', alpha=0.8, linewidth=2)
sns.kdeplot(predictions_df['severe'], shade=True, label='Severe', alpha=0.8, linewidth=2)

# Plot settings
plt.title('Density Plot', fontweight='bold', fontsize=15)
plt.xlabel('Probability', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=14)


plt.tight_layout()
plt.show();
```
